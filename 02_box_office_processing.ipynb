```python
import pandas as pd
```

```python
df_og = pd.read_csv("output_data/2021_03_04_DomesticBoxOfficeMojo_2010-2021.csv")
```

```python
df_og
```

```python
# create list of years
years = df_og["Year"].unique().tolist()
print(years)
```

```python
# create list of ranks, top 100
high_rank = df_og["Rank"].unique().tolist()[0:100]
print(high_rank)
```

```python
# create new dataframe with only the top 25 grossing films from each year
df_top100 = df_og.loc[(df_og["Rank"].isin(high_rank)) & (df_og["Year"].isin(years))]
```

```python
# reset index
df_top100 = df_top100.reset_index(drop=True)
```

```python
# test results
df_top100.loc[df_top100["Year"] == 2016]
```

```python
df_top100.to_csv("output_data/Top100_Movies_2010-2021.csv", index=False, header=True)
```
