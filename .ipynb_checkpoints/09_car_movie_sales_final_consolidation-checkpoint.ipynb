```python
import pandas as pd
```

```python
# read in each year of data
df_10 = pd.read_csv('output_data/raw/final/car_mov_sales_2010.csv')
df_11 = pd.read_csv('output_data/raw/final/car_mov_sales_2011.csv')
df_12 = pd.read_csv('output_data/raw/final/car_mov_sales_2012.csv')
df_13 = pd.read_csv('output_data/raw/final/car_mov_sales_2013.csv')
df_14 = pd.read_csv('output_data/raw/final/car_mov_sales_2014.csv')

df_15 = pd.read_csv('output_data/raw/final/car_mov_sales_2015.csv')
df_16 = pd.read_csv('output_data/raw/final/car_mov_sales_2016.csv')
df_17 = pd.read_csv('output_data/raw/final/car_mov_sales_2017.csv')
df_18 = pd.read_csv('output_data/raw/final/car_mov_sales_2018.csv')
df_19 = pd.read_csv('output_data/raw/final/car_mov_sales_2019.csv')
df_20 = pd.read_csv('output_data/raw/final/car_mov_sales_2020.csv')
```

```python
# append 2011 dataframe to 2010
df_master = df_10.append(df_11)
```

```python
# iterate through list of dataframes, appending them one at a time
df_list = [df_12, df_13, df_14, df_15, df_16, df_17, df_18, df_19, df_20]

for df in df_list:
    df_master = df_master.append(df)
```

```python
# reorder columns and lose 'Movie Presence'
df_master = df_master[['Year', 'Vehicle Original', 'Vehicle', 'Sales', '% Change', 'Change',
                       'Car', 'Stars', 'Highest Star', 'Strong Appearance', 'Weak Appearance', 'Total Appearance',
                       'Release', 'Rank', 'Best Rank', 'Gross', 'Distributor']]
```

```python
# export for review
df_master.to_csv(f"output_data/cars_movies_master.csv", index=False, header=True)
```
