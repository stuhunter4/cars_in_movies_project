```python
import pandas as pd
import numpy as np
import math
```

```python
# read in all of the car sales datasets
clean_2010 = pd.read_csv("output_data/raw/clean_cars_2010.csv")
clean_2011 = pd.read_csv("output_data/raw/clean_cars_2011.csv")
clean_2012 = pd.read_csv("output_data/raw/clean_cars_2012.csv")
clean_2013 = pd.read_csv("output_data/raw/clean_cars_2013.csv")
clean_2014 = pd.read_csv("output_data/raw/clean_cars_2014.csv")

clean_2015 = pd.read_csv("output_data/raw/clean_cars_2015.csv")
clean_2016 = pd.read_csv("output_data/raw/clean_cars_2016.csv")
clean_2017 = pd.read_csv("output_data/raw/clean_cars_2017.csv")
clean_2018 = pd.read_csv("output_data/raw/clean_cars_2018.csv")
clean_2019 = pd.read_csv("output_data/raw/clean_cars_2019.csv")
clean_2020 = pd.read_csv("output_data/raw/clean_cars_2020.csv")
```

# Assemble datasets

```python
# Remove columns for '% Change...'
clean_2010 = clean_2010.drop(['% Change 2010'], axis = 1)
clean_2011 = clean_2011.drop(['% Change 2011'], axis = 1)
clean_2012 = clean_2012.drop(['% Change 2012'], axis = 1)
clean_2013 = clean_2013.drop(['% Change 2013'], axis = 1)
clean_2014 = clean_2014.drop(['% Change 2014'], axis = 1)

clean_2015 = clean_2015.drop(['% Change 2015'], axis = 1)
clean_2016 = clean_2016.drop(['% Change 2016'], axis = 1)
clean_2017 = clean_2017.drop(['% Change 2017'], axis = 1)
clean_2018 = clean_2018.drop(['% Change 2018'], axis = 1)
clean_2019 = clean_2019.drop(['% Change 2019'], axis = 1)
clean_2020 = clean_2020.drop(['% Change 2020'], axis = 1)
```

### First Merge

```python
# first merge with a full outer join to keep all vehicle data from both years
merge10_11 = pd.merge(clean_2010, clean_2011, how="outer", on="Vehicle")
```

```python
# create new column for '2010 Year End' that compares our two lists and keeps only the data
merge10_11["2010 Year End"] = ""

for index, row in merge10_11.iterrows():
    if row[1] == row[4]:
        merge10_11["2010 Year End"][index] = merge10_11["2010 Year End_x"][index]
    else:
        if math.isnan(row[1]):
            merge10_11["2010 Year End"][index] = merge10_11["2010 Year End_y"][index]
        else:
            merge10_11["2010 Year End"][index] = merge10_11["2010 Year End_x"][index]
```

```python
# drop the redundant columns now that data has been merged into a single column
merge10_11 = merge10_11.drop(['2010 Year End_x', '2010 Year End_y'], axis = 1)
```

### Repeat Operations for Multiple Combinations

```python
merge_12 = pd.merge(merge10_11, clean_2012, how="outer", on="Vehicle")
```

```python
# same iteration and conditionals as above to combine data from two columns...
# ...each column with its own missing elements relative to the other
merge_12["2011 Year End"] = ""

for index, row in merge_12.iterrows():
    if row[2] == row[5]:
        merge_12["2011 Year End"][index] = merge_12["2011 Year End_x"][index]
    else:
        if math.isnan(row[2]):
            merge_12["2011 Year End"][index] = merge_12["2011 Year End_y"][index]
        else:
            merge_12["2011 Year End"][index] = merge_12["2011 Year End_x"][index]
```

```python
merge_12 = merge_12.drop(['2011 Year End_x', '2011 Year End_y'], axis = 1)
```

```python
def car_merge(merge_no, clean_df, x, y, year):
    df = pd.merge(merge_no, clean_df, how="outer", on="Vehicle")
    df[f"{year} Year End"] = ""

    for index, row in df.iterrows():
        if row[x] == row[y]:
            df[f"{year} Year End"][index] = df[f"{year} Year End_x"][index]
        else:
            if math.isnan(row[x]):
                df[f"{year} Year End"][index] = df[f"{year} Year End_y"][index]
            else:
                df[f"{year} Year End"][index] = df[f"{year} Year End_x"][index]

    df = df.drop([f'{year} Year End_x', f'{year} Year End_y'], axis = 1)
    return df
```
