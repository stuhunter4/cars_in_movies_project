```python
import pandas as pd
import requests
import numpy as np
from datetime import date
```

```python
# export dataframe of results with today's date
today = date.today()
d_today = today.strftime("%Y_%m_%d")
```

```python
url = 'https://www.goodcarbadcar.net/2014-usa-new-vehicle-sales-figures-by-mode/'
r = requests.get(url)

dfs = pd.read_html(r.text)
```

```python
dfs
```

```python
dfs[0]
```

```python
# slice off dataframe we want using normal indexing
bestcars = dfs[0]
```

```python
#export as csv for use elsewhere
bestcars.to_csv(f"output_data/{d_today}_bestcars2015.csv", index=False, header=True)
```
