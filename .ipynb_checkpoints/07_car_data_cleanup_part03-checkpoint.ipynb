```python
import pandas as pd
import numpy as np
```

```python
# read in aggregated car sales data for final cleanup
data = pd.read_csv("output_data/raw/cars_assembled_01.csv")
```

```python
data
```

```python
# sort alphabetically by Vehicle column
data = data.sort_values("Vehicle")

# reset index
data = data.reset_index(drop=True)

# reorder columns
data = data[['Vehicle', '2009 Year End', '2010 Year End', '2011 Year End', '2012 Year End',
             '2013 Year End', '2014 Year End', '2015 Year End', '2016 Year End', '2017 Year End',
             '2018 Year End', '2019 Year End', '2020 Year End']]

# rename columns
data.columns = ['Vehicle', '2009', '2010', '2011', '2012',
                '2013', '2014', '2015', '2016', '2017',
                '2018', '2019', '2020']
```

```python
data
```

```python
data.to_csv("output_data/Car_Sales_2009-2020.csv", index=False, header=True)
```
