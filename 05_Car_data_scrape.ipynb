```python
import pandas as pd
import requests
import numpy as np
from datetime import date
```

```python
# export dataframe of results with today's date
today = date.today()
d_today = today.strftime("%Y_%m_%d")
```

```python
url = 'https://www.goodcarbadcar.net/december-2018-the-best-selling-vehicles-in-america-every-vehicle-ranked/'
r = requests.get(url)

dfs = pd.read_html(r.text)
```

```python
dfs
```

```python
dfs[0]
```

```python
# slice off dataframe we want using normal indexing
bestcars = dfs[0]
```

```python
bestcars
```

```python
#export as csv for use elsewhere
bestcars.to_csv(f"output_data/{d_today}_bestcars2018.csv", index=False, header=True)
```
