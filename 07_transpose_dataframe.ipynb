```python
import pandas as pd
```

```python
data = pd.read_csv("output_data/Car_Sales_2009-2020.csv")
```

```python
# set the index
data = data.set_index("Vehicle")
```

```python
# remove index heading
data.rename_axis(None, inplace=True)
data
```

```python
# return the transpose of the DataFrame
new_data = data.transpose()
```

```python
# reset index
new_data = new_data.reset_index()
```

```python
# rename Vehicle column to Years
new_data.rename(columns = {'index':'Years'}, inplace = True)
```
